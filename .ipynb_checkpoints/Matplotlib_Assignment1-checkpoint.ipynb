```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
data = [ 1, 10, 2, 7, 5, 12, 11, 4, 12, 4, 2]
data1 = {'unemployment_rate': [6, 5, 10, 15, 9, 11, 20, 5.3, 5.2, 5.2, 5.3]}
df = pd.DataFrame(data1)
plt.plot(data)
plt.plot(df)
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
genres = ['Comedy','Drama','Comedy|Romance','Adventure|Action|Thriller','Comedy|Drama|Romance']
sizes = [ 20,20,20,20,20]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#bbb499']

plt.pie(sizes,labels=genres, autopct='%1.2f%%', colors =colors,  startangle=180, radius=1.5)

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df, data, s= 50, alpha = 0.4) 
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(df, data, s= 50, alpha = 0.4)
plt.title('Title', c = 'red', fontsize=20)
plt.xlabel('df',fontsize = 16)
plt.ylabel('data',fontsize = 16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python

plt.plot(np.arange(5,11), [5,10,15,20,30,35], 'blue', linewidth = 1)
plt.plot(np.arange(5,11), [6,20,23,30,19,22], color = 'red', linewidth = 2)
plt.plot(np.arange(5,11), [7,25,20,25,20,30], color = 'green', linewidth = 3)
plt.legend(['xx','yy', 'pp'], shadow = True, loc = (2))
plt.xlabel('Year')
plt.ylabel('Number of Production')
plt.title('Number of Movies Produced every Year')
plt.show()
```

### Q6: Create subplots with three charts and figsize=20,10.

```python
cc=np.arange(11,22)
cc2=np.arange(50,55)
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 20)

#first
plt.subplot(4,4,1) #(number of rows, number of columns, number of plot)
plt.pie(cc2,labels=genres, autopct='%1.1f%%')
plt.title('Genres');

#second
plt.subplot(4,4,2)
plt.scatter(df,data,cc)
plt.title('Profit vs. Vote Average')
plt.xlabel('Vote Average')
plt.ylabel('Profit'); 

plt.subplot(4,4,3)
plt.plot(np.arange(5,11), [5,10,15,20,30,35], 'blue', linewidth = 1)
plt.plot(np.arange(5,11), [6,20,23,30,19,22], color = 'red', linewidth = 2)
plt.plot(np.arange(5,11), [7,25,20,25,20,30], color = 'green', linewidth = 3)
plt.legend(['xx','yy', 'pp'], shadow = True, loc = (2))
plt.xlabel('Year')
plt.ylabel('Number of Production')
plt.title('Number of Movies Produced every Year');
```
